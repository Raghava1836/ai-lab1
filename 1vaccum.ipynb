```python
import random

# Locations
locations = ["A", "B"]

# Initial environment: randomly clean/dirty
env = {
    "A": random.choice(["Clean", "Dirty"]),
    "B": random.choice(["Clean", "Dirty"])
}

# Agent state
current_location = "A"
performance_score = 0
steps = 4  # number of percept–action cycles

# Logs for final tables
location_log = []   # (location/performance, status, percentage)
percept_log = []    # (percept, action, status)


def performance_to_percentage(score, max_score):
    if max_score == 0:
        return 0
    return int((score / max_score) * 100)


# Main loop: simple reflex agent
for t in range(1, steps + 1):
    # Percept: current location and its status
    status = env[current_location]
    percept = f"Location={current_location}, Status={status}"

    # Action selection
    if status == "Dirty":
        action = "Suck"  # clean
        env[current_location] = "Clean"
        performance_score += 10
        status_after = "Clean"
    else:
        # Move: right for "A", left for "B"
        if current_location == "A":
            action = "Right"
            current_location = "B"
        else:
            action = "Left"
            current_location = "A"
        performance_score += 5
        status_after = env[current_location]

    # Log percept–action–status
    percept_log.append((percept, action, status_after))

# Suppose maximum possible performance for steps is 10 per step
max_possible_score = steps * 10
percentage_score = performance_to_percentage(performance_score, max_possible_score)

# Build location/performance log (here: final snapshot)
location_log.append((
    current_location,          # location
    "Running complete",        # status
    f"{percentage_score}%"     # percentage
))

# --------- PRINT TABLES ---------

# Table 1: (location or performance, status, percentage)
print("TABLE 1: (location/performance, status, percentage)")
print(f"{'Location/Perf':<20}{'Status':<20}{'Percentage':<10}")
for loc, stat, perc in location_log:
    print(f"{loc:<20}{stat:<20}{perc:<10}")

print("\nTABLE 2: (percept, action, status)")
print(f"{'Percept':<40}{'Action':<10}{'Status':<10}")
for percept, action, status in percept_log:
    print(f"{percept:<40}{action:<10}{status:<10}")
```
