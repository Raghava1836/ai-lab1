```python
import heapq

goal = [[1,2,3],
        [4,5,6],
        [7,8,0]]

# Find position of a tile
def find_pos(state, value):
    for i in range(3):
        for j in range(3):
            if state[i][j] == value:
                return i, j

# Heuristic: Manhattan Distance
def heuristic(state):
    dist = 0
    for i in range(3):
        for j in range(3):
            val = state[i][j]
            if val != 0:
                gi, gj = find_pos(goal, val)
                dist += abs(i - gi) + abs(j - gj)
    return dist

# Convert state to tuple (for hashing)
def to_tuple(state):
    return tuple(tuple(row) for row in state)

# Generate neighbors
def get_neighbors(state):
    neighbors = []
    x, y = find_pos(state, 0)
    moves = [(1,0), (-1,0), (0,1), (0,-1)]

    for dx, dy in moves:
        nx, ny = x + dx, y + dy
        if 0 <= nx < 3 and 0 <= ny < 3:
            new_state = [row[:] for row in state]
            new_state[x][y], new_state[nx][ny] = new_state[nx][ny], new_state[x][y]
            neighbors.append(new_state)
    return neighbors

# A* Search
def astar(start):
    pq = []
    heapq.heappush(pq, (heuristic(start), 0, start, []))
    visited = set()

    while pq:
        f, g, state, path = heapq.heappop(pq)

        if state == goal:
            return path + [state]

        visited.add(to_tuple(state))

        for neighbor in get_neighbors(state):
            if to_tuple(neighbor) not in visited:
                h = heuristic(neighbor)
                heapq.heappush(pq, (g + h + 1, g + 1, neighbor, path + [state]))

    return None

# -------- MAIN --------
start = []
print("Enter the initial state (3x3) row wise, use 0 for blank:")
for _ in range(3):
    row = list(map(int, input().split()))
    start.append(row)

solution = astar(start)

if solution:
    print("\nSolution steps:")
    for step in solution:
        for row in step:
            print(row)
        print("-----")
else:
    print("No solution found")
```
