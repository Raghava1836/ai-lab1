```python
class MonkeyBanana:
    def __init__(self, monkey_pos, crate_pos, banana_pos):
        self.monkey_pos = monkey_pos
        self.crate_pos = crate_pos
        self.banana_pos = banana_pos
        self.on_crate = False
        self.has_banana = False
        self.crate_moved = False

    def print_crate(self, found, moved):
        print(f"crate1({self.crate_pos} , {found} , {moved})")

    def perceive_and_act(self):
        # If monkey not at crate
        if self.monkey_pos != self.crate_pos:
            self.print_crate("not found", "not moved")
            self.monkey_pos = self.crate_pos

        # If crate not under banana
        elif self.crate_pos != self.banana_pos:
            self.print_crate("found", "not moved")
            self.crate_pos = self.banana_pos
            self.monkey_pos = self.banana_pos
            self.crate_moved = True
            self.print_crate("found", "moved")

        # If monkey not on crate
        elif not self.on_crate:
            self.print_crate("found", "moved" if self.crate_moved else "not moved")
            self.on_crate = True

        # Grab banana
        elif self.on_crate and self.monkey_pos == self.banana_pos:
            self.print_crate("found", "moved")
            self.has_banana = True

    def run(self):
        while not self.has_banana:
            self.perceive_and_act()


# -------- MAIN --------
monkey_pos = input("Enter Monkey Position: ")
crate_pos = input("Enter Crate Position: ")
banana_pos = input("Enter Banana Position: ")

agent = MonkeyBanana(monkey_pos, crate_pos, banana_pos)
agent.run()

```
